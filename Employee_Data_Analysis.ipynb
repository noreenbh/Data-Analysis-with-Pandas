```python
#Select specific columns from one of the DataFrames.
selected_columns = employee_data[['Experience (Years)', 'Salary']]  
print("\nSelected Columns from Employee Data:\n", selected_columns)
```

```python
# Filtering rows having female data, having experience between 10 to 15 years and their desiqnation is Database Administrator (DBA)
employee_data[(employee_data['Experience (Years)'] >= 10) & (employee_data['Experience (Years)'] <= 15) 
                        & (employee_data['Gender'] == 'F') & (employee_data['Position'] == 'Database Administrator (DBA)')]	
```

```python
# Filtering rows having male data, having experience between 10 to 15 years and their desiqnation is Database Administrator (DBA)
employee_data[(employee_data['Experience (Years)'] >= 10) & (employee_data['Experience (Years)'] <= 15) 
                        & (employee_data['Gender'] == 'M') & (employee_data['Position'] == 'Database Administrator (DBA)')]
```
